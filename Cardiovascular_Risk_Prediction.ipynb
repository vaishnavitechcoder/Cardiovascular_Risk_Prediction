````python
#applying knn imputer
knn_imputer = KNNImputer(n_neighbors=3, weights="uniform",metric='nan_euclidean')
````

````python
#imputing missing values in feature heartRate
cardio_df1['heartRate'] = knn_imputer.fit_transform(cardio_df1[['heartRate']])

# imputeing missing values in feature Glucose_level
cardio_df1['Glucose_level'] = knn_imputer.fit_transform(cardio_df1[['Glucose_level']])

# imputeing missing values in feature Tot_chol_Level
cardio_df1['Tot_chol_Level'] = knn_imputer.fit_transform(cardio_df1[['Tot_chol_Level']])

# imputeing missing values in feature BMI
cardio_df1['BMI'] = knn_imputer.fit_transform(cardio_df1[['BMI']])

# imputeing missing values in feature Cig_Per_Day
cardio_df1['Cig_Per_Day'] = knn_imputer.fit_transform(cardio_df1[['Cig_Per_Day']])
````

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Answer Here.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null hypothesis: There is no significant difference in glucose levels between patients with and without a high risk of cardiovascular disease.

Alternate hypothesis: There is a significant difference in glucose levels between patients with and without a high risk of cardiovascular disease.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import pandas as pd
from scipy.stats import ttest_ind

# Separate the data into two groups based on the 'TenYearCHD' column (0: No CHD risk, 1: CHD risk)
group_0 = cardio_df1[cardio_df1['Ten_Year_Risk'] == 0]['Glucose_level']
group_1 = cardio_df1[cardio_df1['Ten_Year_Risk'] == 1]['Glucose_level']

# Perform the two-sample t-test
t_statistic, p_value = ttest_ind(group_0, group_1)

# Define the significance level (alpha)
alpha = 0.05

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in glucose levels between the groups.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence of a significant difference in glucose levels.")
````

##### Which statistical test have you done to obtain P-Value?

The data is normally distributed, and the variances of the two groups are equal. The two-sample t-test is also a relatively powerful test, so it is likely to be able to detect a significant difference in glucose levels between the two groups

##### Why did you choose the specific statistical test?

 I used a two-sample t-test to obtain the p-value. This is a statistical test that is used to compare the means of two independent groups. In this case, the two groups are patients with and without a high risk of cardiovascular disease.

Alternate hypothesis: There is a significant difference in systolic blood pressure between patients who are on and not on BP medication.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
import pandas as pd
from scipy.stats import ttest_ind


# Separate the data into two groups based on the 'BP_Medication' column (1: On BP meds, 0: Not on BP meds)
group_0 = cardio_df1[cardio_df1['BP_Medication'] == 0]['Systolic_BP']
group_1 = cardio_df1[cardio_df1['BP_Medication'] == 1]['Systolic_BP']

# Perform the two-sample t-test
t_statistic, p_value = ttest_ind(group_0, group_1)

# Define the significance level (alpha)
alpha = 0.05

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference in systolic blood pressure between the groups.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence of a significant difference in systolic blood pressure.")
````

##### Which statistical test have you done to obtain P-Value?


I conducted a two-sample t-test to obtain the p-value. This is a parametric test that is used to compare the means of two independent groups. The two groups in this case are patients who are on and not on BP medication

##### Why did you choose the specific statistical test?

I chose the two-sample t-test because it is a parametric test that is used to compare the means of two independent groups. In this case, the two groups are patients who are on and not on BP medication.

````python
# Handling Outliers & Outlier treatments
# Detect outliers using the Z-score method for numerical columns
# Assuming 'glucose', 'sysBP', and 'diaBP' are numerical columns
from scipy.stats import zscore

# Define the numerical columns to check for outliers
numerical_columns = ['Glucose_level', 'Systolic_BP', 'Diastolic _BP']

# Calculate the Z-scores for the numerical columns
z_scores = cardio_df1[numerical_columns].apply(zscore)

# Set a threshold for the Z-score (e.g., 3 or -3) to identify outliers
threshold = 3

# Find the rows where any of the Z-scores exceed the threshold (outliers)
outliers = cardio_df1[(z_scores > threshold).any(axis=1)]

# Print the outliers
print("Outliers detected:")
print(outliers)

#  Remove outliers (drop rows with outliers
cardio_df1 = cardio_df1.drop(outliers.index)


# Example: Impute 'glucose' outliers with the median
cardio_df1.loc[cardio_df1['Glucose_level'].isin(outliers['Glucose_level']), 'Glucose_level'] = cardio_df1['Glucose_level'].median()
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

Outlier detection: I used the Z-score method to detect outliers. The Z-score is a measure of how far a data point is from the mean, in terms of standard deviations. I set a threshold of 3 to identify outliers, which means that any data point with a Z-score greater than 3 or less than -3 is considered an outlier.
Outlier removal: I removed the outliers from the dataset. This is a simple and effective way to deal with outliers, but it can also be a bit too aggressive. If you remove too many outliers, you may end up removing important data points.
